#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

```python
print(plt.style.available)
```

```python
plt.style.use('seaborn-v0_8')
plt.style.use('_mpl-gallery-nogrid')
```

```python
font = {'family':'sans-serif',
    'weight' : 'regular',
        'size'   : 10}

plt.rc('font', **font)
```

```python
# MEDIAN STEPS BY MONTH_ENCODED 
steps_month = fitbit['Steps'].groupby(fitbit['Months_encoded']).median()
steps_month
```

```python
# MEDIAN STEPS BY WORK_OR_WEEKEND
steps_days = fitbit['Steps'].groupby(fitbit['Work_or_Weekend']).median().sort_values()
steps_days
```

```python
# MEDIAN STEPS BY MONTH_ENCODED VISUAL
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize= (12,4))
plt.sca(axes[0])
steps_month.plot(kind = 'line', linewidth=2)
plt.xlabel('MONTH')
plt.ylabel('MEDIAN STEPS')
plt.xticks(list(range(12+1)),['', 'Jan','Feb','Mar', 'Apr', 'May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])

# MEDIAN STEPS BY WORK_OR_WEEKEND VISUAL
plt.sca(axes[1])
steps_days.plot(kind='bar', alpha=0.5)
plt.xlabel('WORK OR WEEKEND')
plt.ylabel('MEDIAN STEPS')
plt.xticks(list(range(2)),['Weekend', 'Work'])

plt.savefig('../figures/steps_month.png', bbox_inches='tight')
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
fitbit_2 = fitbit[['Minutes Lightly Active','Minutes Very Active','Minutes Sedentary','Steps']]
columns = ['Minutes Lightly Active','Minutes Very Active','Minutes Sedentary','Steps']
fitbit_2
```

```python
fig, axes = plt.subplots(figsize = (16,8), nrows = 1, ncols = (len(columns)-1))
for i in range((len(columns)-1)):
    plt.sca(axes[i])
    plt.scatter(fitbit_2.iloc[:, i], fitbit_2['Steps'], alpha = 0.8)
    plt.xlabel(fitbit_2.iloc[:, i].name)
    plt.ylabel('Steps')
```
